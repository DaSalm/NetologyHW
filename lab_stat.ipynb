5. Визуализировать несколько взаимосвязей между
переменными

price VS engine-size - зависимы, коррелируются

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

np_scaled = min_max_scaler.fit_transform(df[['price', 'engine-size']])
df_norm = pd.DataFrame(np_scaled)
print(df_norm[:5])
```

```python
plt.figure(figsize=(10,5)) # размер графика
# plt.xlim((-1, -0.5))
# plt.ylim((-1, -0.5))
plt.plot(df_norm[0] , df_norm [1],'r.')
```

```python
df_norm.hist(figsize=(5,5))
df_norm.plot(kind="density", layout=(6,5), 
             subplots=True,sharex=False, sharey=False, figsize=(25,25))
plt.show()
```

curb-weight VS highway-mpg - обратно зависимы?

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(-1,1))

np_scaled = min_max_scaler.fit_transform(df[['curb-weight', 'highway-mpg']])
df_norm = pd.DataFrame(np_scaled)
print(df_norm[:5])
```

```python
plt.figure(figsize=(10,5)) # размер графика
# plt.xlim((-1, -0.5))
# plt.ylim((-1, -0.5))
plt.plot(df_norm[0], df_norm[1],'b.')
```

```python
df_norm.hist(figsize=(5,5))
df_norm.plot(kind="density", layout=(6,5), 
             subplots=True,sharex=False, sharey=False, figsize=(25,25))
plt.show()
```
